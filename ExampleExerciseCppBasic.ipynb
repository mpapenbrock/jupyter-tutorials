Generate a Gaussian distribution centered at 2 and with a width of 3 and plot it.
References: 
1) https://root.cern.ch/root/htmldoc/guides/users-guide/ROOTUsersGuide.html#creating-histograms 
2) 

```c++
gRandom = new TRandom3();
TH1D *hGaus = new TH1D("Gaus","Gaus",5,-10,10);
for (int iEntry = 0; iEntry < 100; iEntry++) {
    Double_t value = (gRandom->Gaus(2,3));
    hGaus->Fill(value);
}
TCanvas *can = new TCanvas();
hGaus->Draw();
can->Draw();
```

Fit the above histogram with a Gaussian.
