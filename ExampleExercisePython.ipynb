Plot a Gaussian distribution.

```python
import ROOT
```

```python
hGaus = ROOT.TH1F("histName","histTitle",100,-4,4)
hGaus.FillRandom("gaus")
```

```python
canvas = ROOT.TCanvas("canName","canTitle",800,600)
hGaus.Draw()
canvas.Draw()
```
